```python
def rowSample(df,percent):
    data=df.sample(int(df.shape[0]*percent),replace=True)
    return data

def columnSample(df,percent):
    cols=random.sample(df.columns.tolist()[:-1],int(df.shape[1]*percent))
    return df[cols]

```

```python
df1=rowSample(df,0.4)
df2=columnSample(df,0.4)
```

```python
df1=rowSample(df,0.1)
df2=rowSample(df,0.1)
df3=rowSample(df,0.1)
```
